```python
def plot_correlation_map(df):
    corr = df.corr()
    s , ax = plt.subplots(figsize = (20 , 15))
    cmap = sns.diverging_palette(220 , 10 , as_cmap = True)
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
        )
plot_correlation_map(df_transf)
```

### => MODEL CHOOSING

```python
def apply_model(X, y, model):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
    pipe = make_pipeline(StandardScaler(), model)
    pipe.fit(X_train, y_train)
    y_pred  = pipe.predict(X_test)
    if isinstance(model, LogisticRegression):
        confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
        sns.heatmap(confusion_matrix, annot=True)
        print(metrics.classification_report(y_test, y_pred))
    return round(cross_val_score(pipe, X_train, y_train, cv=5, scoring='accuracy').mean(), 2)
```

```python
X = df_transf.drop(['class'], axis=1)
y = df_transf['class']
algo_scores = {}
```

#### ==> LOGISTIC REGRESSION

```python
model = LogisticRegression()
accuracy = apply_model(X, y, model)
algo_scores['LogisticRegression'] = accuracy
print(f"Accuracy: {accuracy}")
```

knn_scores = []
for k in range(1, 50):
    model = KNeighborsClassifier(k)
    accuracy = apply_model(X, y, model)

```python
model = KNeighborsClassifier(5)
accuracy = apply_model(X, y, model)
algo_scores['KNeighborsClassifier'] = accuracy
```

#### ==> RANDOM FOREST

```python
model = RandomForestClassifier(n_estimators=10)
accuracy = apply_model(X, y, model)
algo_scores['RandomForestClassifier'] = accuracy
print(f"Accuracy: {accuracy}")
```
